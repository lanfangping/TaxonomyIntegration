```python
import os
import json
import pandas as pd
```

```python
file_path = './processed/processed_sent_level.json'
# Step 1: Verify the file path
if not os.path.exists(file_path):
    print("File not found:", file_path)
else:
    print("File found:", file_path)

df = pd.read_json(file_path, lines=True)
df
```

```python
# Define the sampling function
def sample_data(group, n=1, random_state=42):
    return group.sample(n=min(len(group), n), random_state=random_state)

# Number of samples per group
n_samples = 5
random_seed = 41

# Sample n rows per 'domain' and 'label' combination with a random seed
sampled_df = df.groupby(['domain', 'label']).apply(sample_data, n=n_samples, random_state=random_seed).reset_index(drop=True)

# Display the sampled DataFrame
sampled_df
```

```python
# Get the indices of the sampled data
sampled_indices = sampled_df.index

# Drop the sampled data to get the remaining data
remaining_df = df.drop(sampled_indices)

remaining_df = remaining_df.reset_index(drop=True)
```

```python
# Convert DataFrame to JSON Lines format and save to a file
sampled_df.to_json("../sampling/IteraTeR/pilot_study_iterater.json", orient='records', lines=True)
```

```python
n_samples_list = [1, 2, 2]
random_seed = 41

for idx, n_samples in enumerate(n_samples_list):
    # Sample n rows per 'domain' and 'label' combination with a random seed
    sampled_iter_df = sampled_df.groupby(['domain', 'label']).apply(sample_data, n=n_samples, random_state=random_seed).reset_index(drop=True)

    sampled_iter_df.to_json(f"../sampling/IteraTeR/pilot_study_iterater_iter{idx+1}.json", orient='records', lines=True)

    # Get the indices of the sampled data
    sampled_iter_indices = sampled_iter_df.index

    # Drop the sampled data to get the remaining data
    sampled_df = sampled_df.drop(sampled_iter_indices)
    # Reset the index
    sampled_df = sampled_df.reset_index(drop=True)
```

```python
## dataset for comparing group
# Number of samples per group
n_samples = 71
random_seed = 41

# Sample n rows per 'domain' and 'label' combination with a random seed
comparing_df = remaining_df.groupby(['domain', 'label']).apply(sample_data, n=n_samples, random_state=random_seed).reset_index(drop=True)
comparing_df.to_json("../sampling/IteraTeR/comparing_group1_iterater.json", orient='records', lines=True)
```
